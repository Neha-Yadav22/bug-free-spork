    This ATM System project is written in Python. This is a menu driven program.
    In this project we defined a class named ATM in which:
    It contains diffrent attributs they are :
    -> pin
    -> balance
    -> checking_balance
    
    and the class also contains diffrent methods they are:
    
    -> validate()      :- This method is used to validate the pin entered used by the user.

### Diffrent Methods:

####    1. def validate():
        -: This method is used to validate the pin entered by the user.
        -: In this method it takes pin as input from the user two times.
        -: If both inputed pins are equal then the method return True otherwise return False.
    
####    2. def menu():
        -: This method is used to display the menu for user so that the user can interact with the program.
        -: This functions shows diffrent options for user:
            1. Check Balance
            2. Deposit
            3. Withdraw
            4. Transfer Funds
            5. Exit
        -: The user have to input a number i.e. 1,2,3,4 or 5 only so that the program acts accordingly.
    
####    3. def balance():
        -: This method is used to display the current balance of the user.
        
####    4. def deposit():
        -: This method is used to deposit funds to the current user's account.
        -: This method takes the depositing amount as input from the user.
        -: Then add the input amount to the current user account balance.
        -: Then print the final balance of the account.
       
####    5. def withdraw():
        -: This method is used to withdraw funds from the current user's account.
        -: This method takes the withdrawl amount as input from the user.
        -: If sufficient balance present in the account :
            -> Then reduce the inputed withdrawl amount from the current user's account balance.
            -> Othewise, the method outputs the message of insufficient balance.
        -: Then print the final remaining balance of the current user's account.
        
####    6. def tranfer_Funds():
        -: This method is used to transfer some funds from the checking balance to current account balance.
        -: This method takes the transfering amount as input from the user.
        -: If sufficient checking balance present in the account :
            -> Then deposit the inputed transfering amount to the current user's account balance.
            -> Othewise, the method outputs the message of insufficient checking balance.
        -: Then print the final remaining checking balance and account balance of the current user's account.
        
       
        
        
        
        

```python
# imorting modules
import random
import math
```

```python
class ATM():    #using Atm as the class name.
    
    def __init__(self, pin='1234', balance=0):    #constructor
        print('--------------------------------------------')
        print("     -: Welcome to the ATM :-        ")
        print('--------------------------------------------')
        
        while True:
            pin=input("Please Enter your 4 digit pin : ")  #input pin
            print()
            if len(pin)!=4:   #checking pin size
                print("Wrong pin, please enter 4 digit pin only.")
                print()
                continue
            else:
                break
        
        if self.validate(pin):                # checking that the entered pin is valid or not
            
            self.__pin = pin                  # intialize pin variable
            self.__balance = balance          # intialize balance variable
            self.__checking_balance = 10000   # intialize checking balance variable
            self.menu()                       #calling the menu method
        
        else:
            # showing appropriate messages when wrong pin entered              
            print("Worng Pin Entered !!")
            print("Please Restart the System Again")
            print()
    
    def validate(self,pin):  #use to validate pin
        
        pin_copy=input("Enter your 4 digit pin again :") # input pin
        print()
        
        return True if(pin==pin_copy) else False  #return whether both pins are equal are not 
        
    
    def menu(self): #use to display menu
        print("""
            WELCOME TO THE BANK 
        
            Menu:
            1. Check Balance
            2. Deposit
            3. Withdraw
            4. Transfer Funds
            5. Exit
        
        """)
        print()
        
        while True:
            option = input("Enter 1, 2, 3, 4 or 5: ")  #input option 
            print()
            if option not in {'1','2','3','4','5'}:    #checking that input option is valid or not
                print("Invalid Choice!! \n Please Enter : 1, 2, 3, 4, or 5 only!\n")
                print()
                continue
            else:
                if option == '1':            # if 1 select then calling balance method
                    self.balance()
                elif option == '2':          # if 2 select then calling deposit method
                    self.deposit()
                elif option == '3':          # if 3 select then calling withdraw method
                    self.withdraw()
                elif option == '4':          # if 4 select then calling transfer_Funds method
                    self.transfer_Funds() 
                elif option == '5':          # if 5 select then calling recipt method
                    self.recipt()
                    break
    
    def recipt(self):    # use to print the final recipt
        print(f"""
                    PRINTING THE RECIPT...
          
                 ALL THE TRANSACTION ARE COMPLETED                         
              
                    Transaction number       : {str('TR'+str(random.randint(10000, 1000000)))}
                    Account Balance          : {str(self.__balance)+'.00'}  
                    Account Checking Balance : {str(self.__checking_balance)+'.00'}                
              
                   Thanks for using our services 
                      !!Have A Good Day!!
          
          """)
    
    
    def balance(self):  #print the account balance
        print('-'*40)
        print("The Total Account Balance is :",str(self.__balance)+'.00')
        print('-'*40)
        print()
        
    def deposit(self): #deposit the amount to user's account
        
        deposit_amount=int(input("Enter The Amount you want to deposit : ")) #taking input
        print()
        
        self.__balance += deposit_amount  #depositing amount
        self.balance()                    #calling balance method to print the current balance
        
    def withdraw(self):  #withdraw the amount from the user's account
        
        withdraw_amount = int(input("Enter The Amount you want to withdraw : "))  #taking input
        print()
        
        if self.__balance >= withdraw_amount:  #sufficient balance is present or not
            
            self.__balance -= withdraw_amount  # withdrawing the amount from the account 
            
            print("The Withdrwal Amount is : ",str(withdraw_amount)+'.00')  
            print()
            
            self.balance()   #calling balance method to print the current balance
        
        else:
            print("Insufficient Balance")
            print()
    
    def transfer_Funds(self): #transfer the ammount of user's account
        
        transfer_amount = int(input("Enter The Amount you want to Tranfer : "))  #taking input
        print()
        
        if self.__checking_balance >= transfer_amount:  #sufficient balance is present or not
        
            self.__checking_balance -= transfer_amount      #transferring balance
            self.__balance += transfer_amount           #updating account balance
            
            print("The Remaining Checking Balance available is : ",str(self.__checking_balance)+'.00')
            print()
            
            self.balance()  #calling balance method to print the current balance
        
        else:
            print("Not Sufficient Checking Balance available to transfer")
```
